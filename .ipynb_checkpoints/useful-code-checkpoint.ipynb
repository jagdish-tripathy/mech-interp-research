```python
# Code to run trials

cd /workspace/code

# Gemma-2-9B trial 
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-2-9b-it-Q4_K_M.gguf \
  --mode ab \
  --samples 200 | tee error.log

# Gemma-3-12B trial
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-3-12b-it-qat-q4_0.gguf \
  --mode ab \
  --samples 200 | tee error.log

# Gemma-2-9B trial - Score
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-2-9b-it-Q4_K_M.gguf \
  --mode score \
  --samples 200 | tee error.log

# Gemma-3-12B trial - Score
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-3-12b-it-qat-q4_0.gguf \
  --mode score \
  --samples 200 | tee error.log

##################### FULL RUN ########################
# Gemma-2-9B 
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-2-9b-it-Q4_K_M.gguf \
  --mode ab \
  --samples 3000 | tee error.log

# Gemma-3-12B 
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-3-12b-it-qat-q4_0.gguf \
  --mode ab \
  --samples 3000 | tee error.log

# Gemma-2-9B - Score
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-2-9b-it-Q4_K_M.gguf \
  --mode score \
  --samples 3000 | tee error.log

# Gemma-3-12B - Score
python runpod_inference_llama.py \
  --csv /workspace/data/mortgage_bias_dataset.csv \
  --model /workspace/models/gemma-3-12b-it-qat-q4_0.gguf \
  --mode score \
  --samples 3000 | tee error.log


#!/bin/bash
# Main experiment runner with progress tracking

set -e

# Config

cd /workspace/code

CSV="/workspace/data/mortgage_bias_dataset.csv"
MODEL_DIR="/workspace/models"
SCRIPT="/workspace/code/runpod_inference_llama.py"

# PT models only
PT_MODELS=(
    "gemma-3-12b-pt-q4_0.gguf"
)

# Run all PT experiments
for model in "${PT_MODELS[@]}"; do
    for mode in ab score; do
        echo "Running: $model - $mode mode - 3000 samples"
        python runpod_inference_llama.py \
            --csv "$CSV" \
            --model "$MODEL_DIR/$model" \
            --mode "$mode" \
            --samples 3000
    done
done

echo "✅ All PT model experiments complete!"
```

```python
# To check whether the model is correctly loaded.
python << 'EOF'
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

MODEL_PATH = "/workspace/models/gemma-3-12b-it"

# Load with float32
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    device_map="auto",
    torch_dtype=torch.float32,
    trust_remote_code=True
)
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)

# Simple test
prompt = "Hello, how are you?"
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

with torch.no_grad():
    outputs = model(**inputs, output_hidden_states=True)
    h0 = outputs.hidden_states[0]
    print(f"Hidden state 0 has NaN: {torch.isnan(h0).any().item()}")
    print(f"Hidden state 0 min: {h0.min().item()}")
    print(f"Hidden state 0 max: {h0.max().item()}")
    
    # Try generating
    gen = model.generate(**inputs, max_new_tokens=10)
    text = tokenizer.decode(gen[0])
    print(f"\nGenerated: {text}")
EOF
```
